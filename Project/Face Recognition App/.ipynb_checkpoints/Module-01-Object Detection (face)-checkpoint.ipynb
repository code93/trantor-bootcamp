```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read image
img = cv2.imread('./data/Shubham-Birmi.jpg')
# convert into grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# apply haar cascade
haar = cv2.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
faces = haar.detectMultiScale(gray,1.3,4)
```

```python
print(faces)
```

```python
cv2.rectangle(img,(192,165),(192+130,165+130), (0,255,0),3)
plt.imshow(img)
```

```python
cv2.imshow('object-detect',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# crop
face_crop = img[165:165+130, 192:192+130]
plt.imshow(face_crop)
```

```python
cv2.imwrite('./data/male_01.png', face_crop)
```
