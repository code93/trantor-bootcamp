```python
# Resize
# (1) Shrink
# * cv2.INTERAREA
# (2) Enlarge
# * cv2.INTERCUBIC
# * cv2.INTERLINEAR
```

```python
import numpy as np  #numerical python
import cv2  #computer vision
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading image
img = cv2.imread('./data/Shubham-Birmi.jpg') # bgr
# Convert into Grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray.shape)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
# shrink (w=512, h = 512)
# (w=500,h=400)
img_re = cv2.resize(gray,(500,400),cv2.INTER_AREA)
img_re.shape
```

```python
plt.imshow(img_re, cmap='gray')
plt.show()
```

```python
# enlarge (w=2000, h=1000)
img_en = cv2.resize(gray,(2000,1600),cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en, cmap='gray')
plt.show()
```
